# Data Preprocessing

```python
for col in df:
    print(f'unique values in {col}\n is {df[col].unique()}')
```

```python
df.replace({'No internet service':'No'},inplace=True)
df.replace({'No phone service':'No'},inplace=True)
```

```python
def replace(df,features):
    for feature in features:
        df[feature].replace({'Yes':1,'No':0},inplace=True)
```

```python
data=['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport',
      'StreamingTV','StreamingMovies','PaperlessBilling','Churn']
replace(df,data)
df['gender'].replace({'Female':1,'Male':0},inplace=True)
df.head()
```

```python
df=pd.get_dummies(data=df,columns=['InternetService','Contract','PaymentMethod'],dtype=int)
df.head()
```

## Handling imbalance dataset

```python
x=df.drop(['Churn'],axis='columns')
y=df['Churn']
```

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE(sampling_strategy='minority')
X_resample,y_resample=smote.fit_resample(x,y)
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train_re,x_test_re,y_train_re,y_test_re=train_test_split(X_resample,y_resample,random_state=42,test_size=0.2)
```

# Model Training and Evaluation

```python
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report,confusion_matrix
```

```python
model_para = {
    'Logistic_Regression': {
        'model': LogisticRegression(max_iter=1000),
        'params': {
            'C': [1, 5, 10],
            'solver': ['liblinear', 'saga']  # compatible solvers
        }
    },
    'XGBoost': {
        'model': XGBClassifier(eval_metric="logloss"),
        'params': {
            'n_estimators': [100, 200],
            'max_depth': [3, 5, 7],
            'learning_rate': [0.01, 0.1, 0.2],
            'subsample': [0.8, 1]
        }
    }
}
```

```python
best_score=[]
for model,para in model_para.items():
    clf=GridSearchCV(para['model'],para['params'],cv=5)
    clf.fit(x_train_re,y_train_re)
    best_score.append({
        'model':model,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
        
    })
df = pd.DataFrame(best_score,columns=['model','best_score','best_params'])
df
```

```python
df['best_params']
```

```python
## XGBOOST for training n
```

```python
best_params = {
    'learning_rate': 0.1,
    'max_depth': 7,
    'n_estimators': 100,
    'subsample': 1
}
xgb = XGBClassifier(
    learning_rate=best_params['learning_rate'],
    max_depth=best_params['max_depth'],
    n_estimators=best_params['n_estimators'],
    subsample=best_params['subsample'],
    eval_metric="logloss"  
)
```

```python
xgb.fit(x_train_re,y_train_re)
y_predicted=xgb.predict(x_test_re)
CR=classification_report(y_test_re,y_predicted)
print(CR)
```

```python
cm=confusion_matrix(y_test_re,y_predicted)
plt.figure(figsize=(6,4))
sns.heatmap(cm,annot=True,fmt='d')
plt.title("Confusion Matix")
plt.tight_layout()
```

```python
LR=LogisticRegression(solver='liblinear')
LR.fit(x_train_re,y_train_re)
y_predicted=LR.predict(x_test_re)
CR=classification_report(y_test_re,y_predicted)
print(CR)
```

```python
y_predicted=LR.predict(x_test_re)
cm=confusion_matrix(y_test_re,y_predicted)
plt.figure(figsize=(6,4))
sns.heatmap(cm,annot=True,fmt='d')
plt.title("Confusion Matix")
plt.tight_layout()
```

# Final Thought

### Both models are performing really well, but XGBoost has a slight edge with higher precision, recall, and F1-score compared to Logistic Regression. Since churn prediction often benefits from capturing small improvements in accuracy and balance between classes, Iâ€™d go with XGBoost as the final model.
